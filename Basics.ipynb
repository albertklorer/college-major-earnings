# College Major Earnings

Data visualizations based on job outcomes of students who graduated college between 2010 and 2012. 

```python
%matplotlib inline
import matplotlib.pyplot as plt 
import pandas as pd
from pandas.plotting import scatter_matrix
```

```python
# load in data and preview 
recent_grads = pd.read_csv('recent-grads.csv')
recent_grads.head()
```

```python
# generate summary statistics for numeric columns
for column in ['Total', 'Men', 'Women', 'Rank', 'Sample_size', 'Median', 'Low_wage_jobs', 'Full_time', 'Part_time']:
    print(column)
    print(recent_grads[column].describe())
    print()
```

```python
# drop rows with missing columns
raw_data_count = len(recent_grads)
recent_grads = recent_grads.dropna()
cleaned_data_count = len(recent_grads)

print(raw_data_count - cleaned_data_count, 'columns deleted')
```

```python
# sample size v median
ax = recent_grads.plot(x='Sample_size', y='Median', kind='scatter')
ax.set_title('Sample_size vs. Median')
```

```python
# sample size v unemployment rate
ax = recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter')
ax.set_title('Sample_size vs. Unemployment_rate')
```

```python
# full time v median
ax = recent_grads.plot(x='Full_time', y='Median', kind='scatter')
ax.set_title('Full_time vs. Median')
```

```python
# share women v unemployment rate
ax = recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter')
ax.set_title('ShareWomen vs. Unemployment_rate')
```

```python
# men v median
ax = recent_grads.plot(x='Men', y='Median', kind='scatter')
ax.set_title('Men vs. Median')
```

```python
# women v median
ax = recent_grads.plot(x='Women', y='Median', kind='scatter')
ax.set_title('Women vs. Median')
```

```python
# visualize distribtutions
columns = ['Sample_size', 'Median', 'Employed', 'Full_time', 'ShareWomen',
    'Unemployment_rate', 'Men', 'Women']

fig, axs = plt.subplots(len(columns), 1, figsize=(8, 25))
for i in range(len(columns)):
    axs[i].hist(recent_grads[columns[i]], bins=20)
    axs[i].set_title(columns[i])
```

```python
# scatter matrix of sample size and median
scatter_matrix(recent_grads[['Sample_size', 'Median']], figsize=(10,10))
```

```python
# scatter matrix of sample size, median, and unemployment rate
scatter_matrix(recent_grads[['Sample_size', 'Median', 'Unemployment_rate']], figsize=(10,10))
```
